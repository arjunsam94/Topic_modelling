```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('linkminer_betterhelp_com.csv')
df.head()
```

```python
df['Title'] = df['Title'].astype('str')
```

```python
len(df['Title'])
```

```python
import nltk
from nltk.corpus import stopwords

lemmatizer = nltk.stem.WordNetLemmatizer()
```

# 1. Term-document matrix and cleaning the data

```python
processed_coll = []
for doc in df['Title']:
    tokens = nltk.word_tokenize(doc.lower())
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha()]
    tokens = [token for token in tokens if not token in stopwords.words('english')]
    processed_coll.append(' '.join(tokens))
    
print(processed_coll)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 2), min_df=5)
vector_transform = vectorizer.fit_transform(processed_coll)
terms = vectorizer.get_feature_names_out()
vector_transform_array = vector_transform.toarray()
```

```python
output_df = pd.DataFrame(vector_transform_array, columns=terms)
output_df
```

# 2. LDA model to extract topics. Topics = 6

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_title = LatentDirichletAllocation(n_components=6).fit(vector_transform)
```

```python
title_topic_distribution = lda_title.transform(vector_transform)
```

# 3. Topic distribution and top 2 topics of first 10 page titles for testing

```python
first_10_title = title_topic_distribution[0:10]
print(first_10_title)
```

```python
top_2_topics_titles = np.argsort(-first_10_title)[:, :2]
for i, topics in enumerate(top_2_topics_titles):
    print(f"Title {i}: Top 2 topics are {topics[0]} and {topics[1]}")
```

# 4. Top terms for each topic

```python
for topic_idx, topic in enumerate(lda_title.components_):
    print("Topic %d:" % (topic_idx))
    print(" ".join([terms[i] for i in topic.argsort()[:-6-1:-1]]))
```
